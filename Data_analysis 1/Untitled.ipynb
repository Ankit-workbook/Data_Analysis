```python
# Columns representing investment preferences
preference_cols = [
    'Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds',
    'Fixed_Deposits', 'PPF', 'Gold'
]

# Average preference per option
avg_preferences = df[preference_cols].mean().sort_values()

# Bar plot of average ranks (lower = more preferred)
plt.figure(figsize=(8,5))
sns.barplot(x=avg_preferences.values, y=avg_preferences.index, palette="viridis")
plt.title("Average Preference Ranking (1 = Most Preferred)")
plt.xlabel("Average Rank")
plt.ylabel("Investment Option")
plt.show()
```

This bar chart shows the average preference ranking for various investment options:

Insights:

Mutual Funds and Equity Market are the most preferred, with the lowest average ranks.

Gold is also popular among investors.

PPF and Debentures are ranked lower, indicating they are less preferred.

Government Bonds and Fixed Deposits fall somewhere in the middle.

```python
plt.figure(figsize=(10,6))
corr_matrix = df[preference_cols].corr()
sns.heatmap(corr_matrix, annot=True,  cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Investment Preferences")
plt.show()
```

Strong Positive Correlation between:

Mutual_Funds & Equity_Market (investors who like mutual funds also prefer equity)

PPF & Government_Bonds (conservative, low-risk instruments)

Negative Correlation between:

Gold & PPF/FDs in some cases (risk hedging behavior)

 Insight: Investors tend to group choices by risk profiles. For example:

Aggressive investors favor Mutual Funds + Equity.

Conservative investors lean toward PPF + Bonds.

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Expect", y="age", data=df, palette="Set3")
plt.title("Age vs Expected Return")
plt.xlabel("Expected Return")
plt.ylabel("Age")
plt.show()
```

Younger investors are more likely to expect 20%-30% returns.

A few older investors (30+) show slightly more conservative expectations.

 Insight: As expected, younger investors are more return-aggressive, while older investors begin to temper their expectations slightly.
